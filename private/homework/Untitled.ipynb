```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/paper.mplstyle")
```

```python
x = np.linspace(-10,10,1000)
sigma = np.linspace(1,5,5)
for sig in sigma:
    y = norm(loc=0,scale=sig).pdf(x)
    plt.plot(x,y,label=r'$\sigma={:.0f}$'.format(sig))
plt.xlabel('x')
plt.ylabel(r'normal(x,$\sigma$)')
plt.legend()
plt.savefig('normal_dist.pdf',bbox_inches='tight')
```
