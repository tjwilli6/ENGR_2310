```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def get_xy(npoints,l=1):
    x = []
    y = []
    for i in range(npoints):
        x.append( random.uniform(-l,l) )
        y.append( random.uniform(-1,l) )
    return x,y
```

```python
x,y = get_xy(1000,1)
plt.scatter(x,y)
```

```python
radius = 0.5
tot = 0
for i in range(len(x)):
    if x[i]**2 + y[i]**2 < radius**2:
        tot+=1
area = tot / len(x) * 4
```
