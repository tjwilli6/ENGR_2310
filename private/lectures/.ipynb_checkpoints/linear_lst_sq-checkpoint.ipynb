```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/presentation.mplstyle")
```

## Create Fake Data

```python
x = np.linspace(1,10,10)
ytrue = 3 * x + 5
ymeas = ytrue
plt.scatter(x,ymeas)
```

### Solve for slope and intercept

```python
xmean = x.mean()
ymean = ymeas.mean()
num = 0
denom = 0
for i in range(x.size):
    num += (x[i]-xmean) * (ymeas[i]-ymean)
    denom += (x[i]-xmean)**2
slope = num / denom

intercept = ymean - slope * xmean
print(slope,intercept)
```

```python
#Plot
xplot = np.linspace(x.min(),x.max(),100)
yplot = slope * xplot + intercept
plt.scatter(x,ymeas)
plt.plot(xplot,yplot,c='green')
```

### Create fake data with noise

```python
x = np.linspace(1,10,10)
ytrue = 3 * x + 5
ymeas = ytrue * np.random.normal(1,0.1,size=x.size)
plt.scatter(x,ymeas)
```

```python
xmean = x.mean()
ymean = ymeas.mean()
num = 0
denom = 0
for i in range(x.size):
    num += (x[i]-xmean) * (ymeas[i]-ymean)
    denom += (x[i]-xmean)**2
slope = num / denom

intercept = ymean - slope * xmean
print(slope,intercept)
```

```python
#Plot
xplot = np.linspace(x.min(),x.max(),100)
yplot = slope * xplot + intercept
plt.scatter(x,ymeas)
plt.plot(xplot,yplot,c='green')
```
