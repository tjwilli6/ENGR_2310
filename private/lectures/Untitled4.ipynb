```python
import numpy as np
import matplotlib.pyplot as plt

G = 9.8
rho = 1.2e3
A = 2
C = 1
mass = 1100

def dvdt(v,C,A,rho,mass):
    return -C * A * rho * np.abs(v) * v /mass

vi = 100
yi = 0
dt = 0.01
ti = 0
tf = 10
t = np.arange(ti,tf,dt)
v = np.zeros_like(t) 
y = np.zeros_like(t)
v[0] = vi
y[0] = yi
for i in range(1,t.size):
    vprime = dvdt(v[i-1],C,A,rho,mass)
    v[i] = v[i-1] + vprime * dt
    y[i] = y[i-1] + v[i] * dt
plt.plot(t,y)
```
