```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

```python
#Generate N many random numbers
#(Old way)
random_numbers = []
N = 100
for i in range(N):
    random_numbers.append(random.uniform(0,100))
print(random_numbers)
```

```python
N = 100
random_numbers = np.random.uniform(0,100,size=N)
print(random_numbers)
```

```python
#Average of array of random numbers
avg = sum(random_numbers) / len(random_numbers)
print(avg)

avg = random_numbers.mean()
```

```python
#average doesnt tell the whole story
#avg = (0,40) is the same as average of (19,21)
#The quantity that is different is the standard deviation
#On average, how far away from the mean is each point?
```

```python
#Old way
avg = sum(random_numbers) / len(random_numbers)
mean_dist = []
for value in random_numbers:
    sq_dist = (value - avg)**2
    mean_dist.append(sq_dist)
var = sum(mean_dist) / (len(mean_dist)) 
std = math.sqrt(var)
print(std)
```

```python
#New way
std = random_numbers.std()
print(std)
```

```python
x1 = np.random.uniform(-50,50,100)
x2 = np.random.uniform(-10,10,100)
print(x1.mean(),x1.std())
print(x2.mean(),x2.std())
```
